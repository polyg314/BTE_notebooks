```python
import gene_symptoms_question_functions as gsf
import importlib
import requests
importlib.reload(gsf)
import pandas as pd

from biothings_explorer.user_query_dispatcher import FindConnection
from biothings_explorer.hint import Hint
ht = Hint()
```

```python
# get disease 
xp = ht.query("Xeroderma Pigmentosum")['Disease'][0]
xp
```

```python
# get disease symptoms
xp_symptoms_list = gsf.get_disease_symptoms("Xeroderma Pigmentosum")
xp_symptoms = xp_symptoms_list[0]
xp_symptom_hpids = xp_symptoms_list[1]
xp_symptom_dict = xp_symptoms_list[2]
xp_symptoms
xp_symptom_hpids
```

```python
xp_symptom_dict = gsf.get_symtpom_prevalence(xp_symptom_dict) 
xp_symptoms_df = pd.DataFrame.from_dict(xp_symptom_dict, orient='index').sort_values(by=['edges_out_count'])
xp_symptoms_df = pd.concat([xp_symptoms_df[xp_symptoms_df["frequency"] == "Very frequent"],
               xp_symptoms_df[xp_symptoms_df["frequency"] == "Frequent"],
               xp_symptoms_df[xp_symptoms_df["frequency"] == "Occasional"],
               xp_symptoms_df[xp_symptoms_df["frequency"] == "Rare"],
               xp_symptoms_df[xp_symptoms_df["frequency"] == "Unknown"]
              ])
xp_symptoms_df
```

```python
xp_disease_to_genes_dict = gsf.get_disease_to_gene_results(xp)
```

```python
xp_disease_to_genes_dict['sorted_disease_to_genes']
```

```python
# get results using intermediates 
# print("running disease -> any intermediate -> gene")
node_type_list = (['Gene', 'SequenceVariant', 'ChemicalSubstance', 'Disease', 
                'MolecularActivity', 'BiologicalProcess', 'CellularComponent', 
                'Pathway', 'AnatomicalEntity', 'PhenotypicFeature'])
```

```python
xp_disease_all_nodes_genes = gsf.predict_many(xp, node_type_list, 'Gene')

# Save an object to a file
%store xp_disease_all_nodes_genes
```

```python
%store -r xp_disease_to_node_to_genes_dict
# combine genes from one and two step processes then get rid of any duplicates
xp_top_genes_list = xp_disease_to_genes_dict["disease_to_genes_list"] + xp_disease_to_node_to_genes_dict["top_related_genes_to_disease"]
xp_top_genes_list = list(dict.fromkeys(xp_top_genes_list))
# top_genes_list
```

```python
xp_relevant_genes_to_symptoms_df = gsf.determined_genes_to_symptoms(xp_top_genes_list, xp_symptoms)
# relevant_genes_to_symptoms_df
```

```python
%store xp_relevant_genes_to_symptoms_df
```

```python
%store -r xp_relevant_genes_to_symptoms_df
xp_relevant_genes_list = list(dict.fromkeys(list(xp_relevant_genes_to_symptoms_df["input"])))
# relevant_genes_list
```

```python
xp_gene_to_symptom_pub_counts = gsf.get_gene_to_symptom_publication_counts(xp_relevant_genes_to_symptoms_df)
# gene_to_symptom_pub_counts
```

```python
xp_causes_dict = gsf.create_causes_dict(xp_relevant_genes_to_symptoms_df)
# causes_dict
```
