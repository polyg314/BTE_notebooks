## Get disease symptoms

```python
# get disease symptoms
xp_symptoms_list = gsf.get_disease_symptoms(disease_name)
xp_symptoms = xp_symptoms_list[0]
xp_symptom_hpids = xp_symptoms_list[1]
xp_symptom_dict = xp_symptoms_list[2]
xp_symptoms
xp_symptom_hpids
```

## Get Symptom Prevalence - based off edges out counts

```python
xp_symptom_dict = gsf.get_symtpom_prevalence(xp_symptom_dict, disease_name) 
xp_symptoms_df = pd.DataFrame.from_dict(xp_symptom_dict, orient='index').sort_values(by=['edges_out_count'])
xp_symptoms_df = pd.concat([xp_symptoms_df[xp_symptoms_df["frequency"] == "Very frequent"],
               xp_symptoms_df[xp_symptoms_df["frequency"] == "Frequent"],
               xp_symptoms_df[xp_symptoms_df["frequency"] == "Occasional"],
               xp_symptoms_df[xp_symptoms_df["frequency"] == "Rare"],
               xp_symptoms_df[xp_symptoms_df["frequency"] == "Unknown"]
              ])
```

## Calculate Individual Symptom Scores

```python
individual_symptom_scores = [];
for index,x in xp_symptoms_df.iterrows():
    if(x["frequency"] == "Very frequent"):
        score = 20 / math.sqrt(int(x["edges_out_count"]))
    elif(x["frequency"] == "Frequent"):
        score = 15 / math.sqrt(int(x["edges_out_count"]))
    elif(x["frequency"] == "Occasional"):
        score = 10 / math.sqrt(int(x["edges_out_count"]))
    elif(x["frequency"] == "Rare"):
        score = 5 / math.sqrt(int(x["edges_out_count"]))
    elif(x["frequency"] == "Unknown"):
        score = 5 / math.sqrt(int(x["edges_out_count"]))
    individual_symptom_scores.append(score)

# print(len(individual_symptom_scores))
xp_symptoms_df["ISS"] = individual_symptom_scores
```
